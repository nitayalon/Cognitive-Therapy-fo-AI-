### Inspect Unified Datasets

```python
def load_all_task_data():
    """
    Load data from all tasks in the selected experiment and create unified datasets.
    
    Returns:
        dict: Dictionary containing:
            - 'raw_data': Raw data per task
            - 'training_unified': Combined training data
            - 'test_unified': Combined test data
    """
    all_data = {}
    training_data_list = []
    test_data_list = []
    
    print(f"Loading data from all {num_tasks} tasks...")
    print(f"Experiment: {EXPERIMENT_DIR}")
    print("-" * 80)
    
    for task_id in range(num_tasks):
        game_name, opp_range, opp_probs = TRAINING_CONDITIONS[task_id]
        print(f"\nTask {task_id}: {game_name} ({opp_range})")
        
        task_path = find_task_path(task_id)
        if not task_path:
            print(f"  ⚠ Skipping - path not found")
            continue
        
        task_data = {
            'task_id': task_id,
            'game': game_name,
            'opp_range': opp_range,
            'opp_probs': opp_probs,
            'task_path': task_path
        }
        
        # Load training data
        training_df = load_training_data(task_path)
        if training_df is not None:
            # Add metadata columns
            training_df['task_id'] = task_id
            training_df['game'] = game_name
            training_df['opp_range'] = opp_range
            task_data['training'] = training_df
            training_data_list.append(training_df)
            print(f"  ✓ Training: {len(training_df)} records")
        else:
            print(f"  ⚠ Training data not found")
        
        # Load test results
        test_results = load_test_results(task_path)
        if test_results:
            task_data['test_results'] = test_results
            
            # Extract test summaries
            if 'evaluation_summaries' in test_results:
                for cond_name, summary in test_results['evaluation_summaries'].items():
                    if isinstance(summary, dict):
                        test_row = {
                            'task_id': task_id,
                            'game': game_name,
                            'opp_range': opp_range,
                            'test_condition': cond_name,
                            'mean_reward': summary.get('mean_reward', np.nan),
                            'mean_cooperation_rate': summary.get('mean_cooperation_rate', np.nan),
                        }
                        # Normalize reward
                        if not np.isnan(test_row['mean_reward']):
                            test_row['mean_reward_norm'] = normalize_reward(
                                test_row['mean_reward'], game_name
                            )
                        test_data_list.append(test_row)
                print(f"  ✓ Test results: {len(test_results['evaluation_summaries'])} conditions")
        else:
            print(f"  ⚠ Test results not found")
        
        all_data[task_id] = task_data
    
    # Create unified DataFrames
    training_unified = pd.concat(training_data_list, ignore_index=True) if training_data_list else pd.DataFrame()
    test_unified = pd.DataFrame(test_data_list) if test_data_list else pd.DataFrame()
    
    print("\n" + "=" * 80)
    print("DATA LOADING COMPLETE")
    print("=" * 80)
    print(f"Tasks loaded: {len(all_data)}/{num_tasks}")
    print(f"Training records: {len(training_unified):,}")
    print(f"Test records: {len(test_unified):,}")
    
    return {
        'raw_data': all_data,
        'training_unified': training_unified,
        'test_unified': test_unified
    }


# Load all data
unified_data = load_all_task_data()
```

```python
def verify_task_data(task_id, verbose=True):
    """
    Verify data integrity for a specific task.
    
    Args:
        task_id (int): Task ID to verify
        verbose (bool): Print detailed status
        
    Returns:
        dict: Verification status with issues list
    """
    game_name, opp_range, opp_probs = TRAINING_CONDITIONS[task_id]
    
    status = {
        'task_id': task_id,
        'game': game_name,
        'opp_range': opp_range,
        'path_found': False,
        'config_found': False,
        'training_data_found': False,
        'test_results_found': False,
        'issues': []
    }
    
    # Find task path
    task_path = find_task_path(task_id)
    if not task_path:
        status['issues'].append("Task directory not found")
        return status
    
    status['path_found'] = True
    status['task_path'] = str(task_path)
    
    # Check experiment config
    config = load_experiment_config(task_path)
    if config:
        status['config_found'] = True
    else:
        status['issues'].append("Missing experiment_config.json")
    
    # Check training data
    training_df = load_training_data(task_path)
    if training_df is not None:
        status['training_data_found'] = True
        status['num_training_records'] = len(training_df)
    else:
        status['issues'].append("Missing training log CSV")
    
    # Check test results
    test_results = load_test_results(task_path)
    if test_results:
        status['test_results_found'] = True
        if 'evaluation_summaries' in test_results:
            status['num_test_conditions'] = len(test_results['evaluation_summaries'])
        else:
            status['issues'].append("Test results missing evaluation_summaries")
    else:
        status['issues'].append("Missing test results pickle")
    
    # Check for checkpoints
    checkpoint_dir = task_path / "checkpoints"
    if checkpoint_dir.exists():
        checkpoints = list(checkpoint_dir.glob("*.pth"))
        status['num_checkpoints'] = len(checkpoints)
        if len(checkpoints) == 0:
            status['issues'].append("No checkpoint files found")
    else:
        status['issues'].append("Checkpoints directory missing")
    
    if verbose:
        print(f"\nTask {task_id} ({game_name}, {opp_range}):")
        if status['path_found']:
            print(f"  ✓ Path found: {task_path.name}")
        if status['config_found']:
            print(f"  ✓ Config found")
        if status['training_data_found']:
            print(f"  ✓ Training data: {status['num_training_records']} records")
        if status['test_results_found']:
            print(f"  ✓ Test results: {status.get('num_test_conditions', '?')} conditions")
        if status['issues']:
            print(f"  ⚠ Issues: {', '.join(status['issues'])}")
    
    return status


# Verify all tasks in selected experiment
print("=" * 80)
print(f"VERIFYING DATA INTEGRITY FOR ALL {num_tasks} TASKS")
print(f"Experiment: {EXPERIMENT_DIR}")
print("=" * 80)

verification_results = {}
for task_id in range(num_tasks):
    verification_results[task_id] = verify_task_data(task_id, verbose=True)

# Summary
print("\n" + "=" * 80)
print("VERIFICATION SUMMARY")
print("=" * 80)
tasks_with_issues = sum(1 for v in verification_results.values() if v['issues'])
print(f"Tasks with complete data: {num_tasks - tasks_with_issues}/{num_tasks}")
print(f"Tasks with issues: {tasks_with_issues}/{num_tasks}")

if tasks_with_issues > 0:
    print("\nTasks with issues:")
    for task_id, status in verification_results.items():
        if status['issues']:
            print(f"  Task {task_id}: {', '.join(status['issues'])}")
else:
    print(f"\n✓ All {num_tasks} tasks have complete data!")
```

### Verify Data Integrity for All Tasks

```python
def find_task_path(task_id):
    """
    Find the directory path for a specific task ID in the generalization matrix experiment.
    
    Args:
        task_id (int): Task ID from 0 to 14
        
    Returns:
        Path or None: Path to task directory if found, None otherwise
    """
    exp_dir = EXPERIMENTS_DIR / EXPERIMENT_DIR
    if not exp_dir.exists():
        return None
    
    # Look for task directories matching pattern: generalization_matrix_task_{task_id}_*
    task_dirs = list(exp_dir.glob(f"generalization_matrix_task_{task_id}_*"))
    
    if not task_dirs:
        return None
    
    # If multiple directories found (e.g., from reruns), use the most recent one
    task_dirs.sort(key=lambda p: p.name, reverse=True)
    return task_dirs[0]


def normalize_reward(reward, game_name):
    """
    Min-max normalize reward to [0, 1] based on game's payoff range.
    
    Args:
        reward (float): Raw reward value
        game_name (str): Name of the game
        
    Returns:
        float: Normalized reward in [0, 1]
    """
    min_r, max_r = GAME_PAYOFF_RANGES[game_name]
    if max_r == min_r:
        return 0.5
    return np.clip((reward - min_r) / (max_r - min_r), 0, 1)


def load_training_data(task_path):
    """
    Load training CSV log for a task and aggregate by epoch.
    
    Args:
        task_path (Path): Path to task directory
        
    Returns:
        pd.DataFrame or None: Epoch-aggregated training data if found, None otherwise
    """
    csv_path = task_path / "checkpoints" / "detailed_training_logs" / "detailed_training_log.csv"
    if not csv_path.exists():
        return None
    try:
        # Load detailed step-by-step data
        detailed_df = pd.read_csv(csv_path)
        
        # Aggregate by epoch to create episode-level statistics
        epoch_stats = detailed_df.groupby('epoch').agg({
            'agent_reward': ['mean', 'std', 'min', 'max', 'count'],
            'policy_prob_cooperate': 'mean',
            'total_loss': 'mean',
            'rl_loss': 'mean',
            'opponent_policy_loss': 'mean',
            'alpha': 'mean',
            'value_estimate': 'mean',
            'advantage': ['mean', 'std'],
            'gradient_norm': 'mean'
        }).reset_index()
        
        # Flatten multi-level columns
        epoch_stats.columns = ['_'.join(col).strip('_') if col[1] else col[0] 
                               for col in epoch_stats.columns.values]
        
        # Rename to match expected column names
        epoch_stats = epoch_stats.rename(columns={
            'epoch': 'episode',
            'agent_reward_mean': 'avg_reward',
            'agent_reward_std': 'reward_std',
            'agent_reward_min': 'reward_min',
            'agent_reward_max': 'reward_max',
            'agent_reward_count': 'steps_per_episode',
            'policy_prob_cooperate_mean': 'avg_cooperation',
            'total_loss_mean': 'avg_loss',
            'rl_loss_mean': 'avg_rl_loss',
            'opponent_policy_loss_mean': 'avg_opp_loss',
            'alpha_mean': 'avg_alpha',
            'value_estimate_mean': 'avg_value',
            'advantage_mean': 'avg_advantage',
            'advantage_std': 'advantage_std',
            'gradient_norm_mean': 'avg_gradient_norm'
        })
        
        return epoch_stats
        
    except Exception as e:
        print(f"Error loading training data from {csv_path}: {e}")
        return None


def load_test_results(task_path):
    """
    Load test results pickle for a task.
    
    Args:
        task_path (Path): Path to task directory
        
    Returns:
        dict or None: Test results if found, None otherwise
    """
    results_dir = task_path / "results"
    if not results_dir.exists():
        return None
    
    pkl_files = list(results_dir.glob("*.pkl"))
    if not pkl_files:
        return None
    
    try:
        with open(pkl_files[0], 'rb') as f:
            return pickle.load(f)
    except Exception as e:
        print(f"Error loading test results from {pkl_files[0]}: {e}")
        return None


def load_experiment_config(task_path):
    """
    Load experiment configuration JSON.
    
    Args:
        task_path (Path): Path to task directory
        
    Returns:
        dict or None: Config if found, None otherwise
    """
    config_path = task_path / "experiment_config.json"
    if not config_path.exists():
        return None
    try:
        with open(config_path, 'r') as f:
            return json.load(f)
    except Exception as e:
        print(f"Error loading config from {config_path}: {e}")
        return None


print("✓ Data loading helper functions defined")
```

```python
def load_all_task_repetitions(experiment_dir_name, task_id):
    """Load all repetitions for a specific task."""
    experiment_path = Path("..") / experiment_dir_name
    task_pattern = f"generalization_matrix_task_{task_id}_*"
    matching_dirs = sorted(experiment_path.glob(task_pattern))
    
    all_runs = []
    for task_dir in matching_dirs:
        try:
            data = {}
            
            # Load training history CSV
            csv_path = task_dir / "training_history.csv"
            if csv_path.exists():
                data['training_history'] = pd.read_csv(csv_path)
            
            # Load test results pickle
            pkl_path = task_dir / "test_results.pkl"
            if pkl_path.exists():
                with open(pkl_path, 'rb') as f:
                    data['test_results'] = pickle.load(f)
            
            # Load checkpoints info
            checkpoints_dir = task_dir / "checkpoints"
            if checkpoints_dir.exists():
                data['checkpoints'] = list(checkpoints_dir.glob("*.pth"))
            
            data['task_dir'] = task_dir
            all_runs.append(data)
        except Exception as e:
            print(f"Warning: Failed to load {task_dir.name}: {e}")
    
    return all_runs

# Example: Load all repetitions for a task with multiple runs
if repetitions and max(repetitions.values()) > 1:
    # Find first task with multiple runs
    multi_run_task = next(task_id for task_id, count in repetitions.items() if count > 1)
    print(f"\nLoading all repetitions for Task {multi_run_task}...")
    example_runs = load_all_task_repetitions(EXPERIMENT_DIR, task_id=multi_run_task)
    print(f"Found {len(example_runs)} runs for Task {multi_run_task}")
    for i, run_data in enumerate(example_runs):
        print(f"  Run {i+1}: {run_data['task_dir'].name}")
```

### 6.3 Aggregate Metrics with Confidence Intervals

Calculate mean, std, and 95% CI for metrics across repetitions using scipy.stats.

```python
from scipy import stats

def aggregate_training_metrics(all_runs, metric_name='avg_reward'):
    """Aggregate training metrics across multiple runs with confidence intervals."""
    # Find common episodes across all runs
    all_episodes = []
    for run_data in all_runs:
        if 'training_history' in run_data and metric_name in run_data['training_history'].columns:
            episodes = run_data['training_history']['episode'].values
            all_episodes.append(episodes)
    
    if not all_episodes:
        return None
    
    # Use the minimum common length
    min_length = min(len(eps) for eps in all_episodes)
    common_episodes = all_episodes[0][:min_length]
    
    # Collect metric values for each episode
    metric_values = []
    for run_data in all_runs:
        if 'training_history' in run_data and metric_name in run_data['training_history'].columns:
            values = run_data['training_history'][metric_name].values[:min_length]
            metric_values.append(values)
    
    metric_values = np.array(metric_values)  # Shape: (n_runs, n_episodes)
    n_runs = len(metric_values)
    
    # Calculate statistics
    means = np.mean(metric_values, axis=0)
    stds = np.std(metric_values, axis=0, ddof=1)  # ddof=1 for sample std
    
    # Calculate 95% confidence intervals using t-distribution
    confidence_level = 0.95
    ci_lower = []
    ci_upper = []
    for i in range(len(means)):
        if n_runs > 1:
            ci = stats.t.interval(confidence_level, n_runs - 1, loc=means[i], scale=stds[i] / np.sqrt(n_runs))
            ci_lower.append(ci[0])
            ci_upper.append(ci[1])
        else:
            ci_lower.append(means[i])
            ci_upper.append(means[i])
    
    return pd.DataFrame({
        'episode': common_episodes,
        'mean': means,
        'std': stds,
        'ci_lower': ci_lower,
        'ci_upper': ci_upper,
        'n_runs': n_runs
    })

def aggregate_test_metrics(all_runs):
    """Aggregate test results across multiple runs with confidence intervals."""
    all_test_results = []
    for run_data in all_runs:
        if 'test_results' in run_data:
            all_test_results.append(run_data['test_results'])
    
    if not all_test_results:
        return None
    
    # Get all opponent types from first run
    opponent_types = list(all_test_results[0].keys())
    
    aggregated = {}
    for opp_type in opponent_types:
        # Collect metrics across runs
        coop_rates = []
        avg_rewards = []
        
        for test_result in all_test_results:
            if opp_type in test_result:
                coop_rates.append(test_result[opp_type]['cooperation_rate'])
                avg_rewards.append(test_result[opp_type]['avg_reward'])
        
        n_runs = len(coop_rates)
        
        # Calculate statistics for cooperation rate
        coop_mean = np.mean(coop_rates)
        coop_std = np.std(coop_rates, ddof=1) if n_runs > 1 else 0
        coop_ci = stats.t.interval(0.95, n_runs - 1, loc=coop_mean, scale=coop_std / np.sqrt(n_runs)) if n_runs > 1 else (coop_mean, coop_mean)
        
        # Calculate statistics for average reward
        reward_mean = np.mean(avg_rewards)
        reward_std = np.std(avg_rewards, ddof=1) if n_runs > 1 else 0
        reward_ci = stats.t.interval(0.95, n_runs - 1, loc=reward_mean, scale=reward_std / np.sqrt(n_runs)) if n_runs > 1 else (reward_mean, reward_mean)
        
        aggregated[opp_type] = {
            'cooperation_rate': {
                'mean': coop_mean,
                'std': coop_std,
                'ci_lower': coop_ci[0],
                'ci_upper': coop_ci[1]
            },
            'avg_reward': {
                'mean': reward_mean,
                'std': reward_std,
                'ci_lower': reward_ci[0],
                'ci_upper': reward_ci[1]
            },
            'n_runs': n_runs
        }
    
    return aggregated

# Test aggregation on example task with multiple runs
if repetitions and max(repetitions.values()) > 1:
    print(f"\nAggregating metrics for Task {multi_run_task}...")
    agg_training = aggregate_training_metrics(example_runs, 'avg_reward')
    agg_test = aggregate_test_metrics(example_runs)
    
    print(f"Training aggregation shape: {agg_training.shape if agg_training is not None else 'None'}")
    print(f"Number of runs: {agg_training['n_runs'].iloc[0] if agg_training is not None else 'N/A'}")
    if agg_test:
        print(f"Test opponents: {list(agg_test.keys())}")
```

### 6.4 Visualize Training Variability

Plot training curves with confidence intervals to visualize variability across repetitions.

```python
def plot_training_with_ci(all_runs, task_id, metric_name='avg_reward', title=None):
    """Plot training curves for all repetitions with mean and 95% CI."""
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Aggregate the data
    agg_data = aggregate_training_metrics(all_runs, metric_name)
    
    if agg_data is None:
        print(f"No data available for {metric_name}")
        return
    
    # Plot 1: Mean with CI shading
    ax = axes[0, 0]
    ax.plot(agg_data['episode'], agg_data['mean'], 'b-', linewidth=2, label='Mean')
    ax.fill_between(agg_data['episode'], agg_data['ci_lower'], agg_data['ci_upper'], 
                     alpha=0.3, color='blue', label='95% CI')
    ax.set_xlabel('Episode')
    ax.set_ylabel(metric_name)
    ax.set_title(f'Mean ± 95% CI (n={agg_data["n_runs"].iloc[0]} runs)')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # Plot 2: All individual runs
    ax = axes[0, 1]
    for i, run_data in enumerate(all_runs):
        if 'training_history' in run_data and metric_name in run_data['training_history'].columns:
            df = run_data['training_history']
            ax.plot(df['episode'], df[metric_name], alpha=0.5, label=f'Run {i+1}')
    ax.set_xlabel('Episode')
    ax.set_ylabel(metric_name)
    ax.set_title('All Individual Runs')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # Plot 3: Standard deviation over time
    ax = axes[1, 0]
    ax.plot(agg_data['episode'], agg_data['std'], 'r-', linewidth=2)
    ax.set_xlabel('Episode')
    ax.set_ylabel('Standard Deviation')
    ax.set_title('Variability Across Runs')
    ax.grid(True, alpha=0.3)
    
    # Plot 4: Coefficient of variation (CV = std/mean)
    ax = axes[1, 1]
    cv = agg_data['std'] / np.abs(agg_data['mean'])
    ax.plot(agg_data['episode'], cv, 'g-', linewidth=2)
    ax.set_xlabel('Episode')
    ax.set_ylabel('Coefficient of Variation')
    ax.set_title('Relative Variability (CV = std/|mean|)')
    ax.grid(True, alpha=0.3)
    
    if title:
        fig.suptitle(title, fontsize=16, y=1.00)
    
    plt.tight_layout()
    plt.show()

# Example: Visualize training variability for task with multiple runs
if repetitions and max(repetitions.values()) > 1:
    game, opponent_range, _ = TRAINING_CONDITIONS[multi_run_task]
    plot_training_with_ci(example_runs, task_id=multi_run_task, metric_name='avg_reward',
                          title=f'Task {multi_run_task}: {game} ({opponent_range}) - Training Variability')
```

### 6.5 Comprehensive Statistical Summary

Load all tasks and create a comprehensive summary showing statistical properties.

```python
def create_comprehensive_summary(experiment_dir_name, num_tasks=15):
    """Create a comprehensive summary of all tasks with statistical metrics."""
    summary_data = []
    
    for task_id in range(num_tasks):
        # Load all repetitions for this task
        all_runs = load_all_task_repetitions(experiment_dir_name, task_id)
        n_runs = len(all_runs)
        
        if n_runs == 0:
            continue
        
        # Get task info
        game, opponent_range, opponent_probs = TRAINING_CONDITIONS[task_id]
        
        # Aggregate final training metrics (last 10 episodes)
        agg_training = aggregate_training_metrics(all_runs, 'avg_reward')
        if agg_training is not None and len(agg_training) > 0:
            final_episodes = agg_training.tail(10)
            final_reward_mean = final_episodes['mean'].mean()
            final_reward_std = final_episodes['std'].mean()
            final_reward_ci_width = (final_episodes['ci_upper'] - final_episodes['ci_lower']).mean()
        else:
            final_reward_mean = final_reward_std = final_reward_ci_width = np.nan
        
        # Aggregate test metrics
        agg_test = aggregate_test_metrics(all_runs)
        if agg_test:
            # Average across all opponent types
            test_coop_means = [v['cooperation_rate']['mean'] for v in agg_test.values()]
            test_reward_means = [v['avg_reward']['mean'] for v in agg_test.values()]
            avg_test_coop = np.mean(test_coop_means)
            avg_test_reward = np.mean(test_reward_means)
        else:
            avg_test_coop = avg_test_reward = np.nan
        
        summary_data.append({
            'task_id': task_id,
            'game': game,
            'opponent_range': opponent_range,
            'opponent_probs': f"[{opponent_probs[0]:.2f}, {opponent_probs[-1]:.2f}]",
            'n_runs': n_runs,
            'final_train_reward_mean': final_reward_mean,
            'final_train_reward_std': final_reward_std,
            'final_train_ci_width': final_reward_ci_width,
            'avg_test_coop': avg_test_coop,
            'avg_test_reward': avg_test_reward
        })
    
    return pd.DataFrame(summary_data)

# Create comprehensive summary
print("Creating comprehensive summary for all tasks...")
summary_df = create_comprehensive_summary(EXPERIMENT_DIR)

# Display summary
print("\n" + "="*100)
print("TASK SUMMARY WITH STATISTICAL PROPERTIES")
print("="*100)
print(summary_df.to_string(index=False))

# Highlight tasks with different statistical power
print("\n" + "="*100)
print("STATISTICAL POWER DISTRIBUTION")
print("="*100)
repetition_counts = summary_df.groupby('n_runs').size()
print(repetition_counts)

print("\nTasks grouped by number of repetitions:")
for n_reps in sorted(summary_df['n_runs'].unique()):
    tasks = summary_df[summary_df['n_runs'] == n_reps]['task_id'].tolist()
    print(f"  {n_reps} repetition(s): Tasks {tasks}")
```

### 6.6 Statistical Power Visualization

Compare confidence interval widths and statistical power across all tasks.

```python
def plot_statistical_power_comparison(summary_df):
    """Visualize statistical power across all tasks."""
    fig, axes = plt.subplots(2, 2, figsize=(16, 10))
    
    # Plot 1: Number of repetitions per task
    ax = axes[0, 0]
    colors = ['red' if n < 3 else 'orange' if n < 5 else 'green' for n in summary_df['n_runs']]
    ax.bar(summary_df['task_id'], summary_df['n_runs'], color=colors, alpha=0.7, edgecolor='black')
    ax.set_xlabel('Task ID')
    ax.set_ylabel('Number of Repetitions')
    ax.set_title('Statistical Power by Task')
    ax.axhline(y=5, color='green', linestyle='--', alpha=0.5, label='Target: 5 runs')
    ax.axhline(y=3, color='orange', linestyle='--', alpha=0.5, label='Minimum: 3 runs')
    ax.legend()
    ax.grid(True, alpha=0.3, axis='y')
    
    # Plot 2: CI width vs number of repetitions
    ax = axes[0, 1]
    valid_data = summary_df.dropna(subset=['final_train_ci_width'])
    scatter = ax.scatter(valid_data['n_runs'], valid_data['final_train_ci_width'], 
                        c=valid_data['task_id'], cmap='viridis', s=100, alpha=0.7, edgecolor='black')
    ax.set_xlabel('Number of Repetitions')
    ax.set_ylabel('95% CI Width (Final Training Reward)')
    ax.set_title('Confidence Interval Width vs Statistical Power')
    plt.colorbar(scatter, ax=ax, label='Task ID')
    ax.grid(True, alpha=0.3)
    
    # Plot 3: Final training reward by game and opponent range
    ax = axes[1, 0]
    for game in GAMES:
        game_data = summary_df[summary_df['game'] == game]
        ax.errorbar(range(len(game_data)), game_data['final_train_reward_mean'], 
                   yerr=game_data['final_train_reward_std'], 
                   marker='o', label=game, capsize=5, linewidth=2)
    ax.set_xlabel('Task Index within Game')
    ax.set_ylabel('Final Training Reward (Mean ± Std)')
    ax.set_title('Training Performance by Game')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # Plot 4: Heatmap of CI widths
    ax = axes[1, 1]
    pivot_data = summary_df.pivot_table(
        values='final_train_ci_width',
        index='game',
        columns='opponent_range',
        aggfunc='mean'
    )
    # Reorder columns
    col_order = ['very_low', 'low', 'mid', 'high', 'very_high']
    pivot_data = pivot_data[[col for col in col_order if col in pivot_data.columns]]
    
    sns.heatmap(pivot_data, annot=True, fmt='.3f', cmap='RdYlGn_r', 
                ax=ax, cbar_kws={'label': 'CI Width'})
    ax.set_xlabel('Opponent Range')
    ax.set_ylabel('Game')
    ax.set_title('Measurement Precision Heatmap\n(Lower = Better)')
    
    plt.tight_layout()
    plt.show()

# Visualize statistical power
plot_statistical_power_comparison(summary_df)
```
